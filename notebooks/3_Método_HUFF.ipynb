```python
import numpy as np
##Cafetería Quetzal##

# Datos de tiendas
tiendas = ["Colonia Moderna", "Tlajomulco", "Zapopis"]

# Atractivo (por ejemplo tamaño en m2)
atractivo = np.array([120, 90, 60])

# Distancia al cliente o zona
distancia = np.array([6, 12, 30])

# Parámetro beta
beta = 3

# Fórmula del método Huff
numerador = atractivo / (distancia ** beta)
probabilidad = numerador / numerador.sum()

# Resultados
for i in range(len(tiendas)):
    print(tiendas[i], "->", round(probabilidad[i], 3))
```
