# BLOQUE 01 – PERFIL DEMOGRÁFICO
## Objetivo
Construir capa consolidada demográfica normalizada por unidad espacial.
## Inputs
- Tipo de Vía.gpkg

## Procesos
- Limpieza
- Homologación de CRS
- Join espacial
- Normalización

## Output
-1_D4C1V1_output_Estructura_vial
- 1_D2C2V1_Tipo de Vía.gpkg


```python
import geopandas as gpd
import pandas as pd
import os
```

```python
# --- 1. CONFIGURACIÓN DE RUTAS ---
ruta_entrada = r"C:\Users\aldov\OneDrive\Escritorio\DAL_LABORATY\data_raw\Vialidades\Tipo de Vía.gpkg"
carpeta_salida = r"C:\Users\aldov\OneDrive\Escritorio\DAL_LABORATY\outputs\1_D4C1V3_Estructura_vial"
nombre_archivo_final = "1_D4C1V3_output_Estructura_vial.gpkg"
```

```python
# Crear carpeta de salida si no existe
if not os.path.exists(carpeta_salida):
    os.makedirs(carpeta_salida)

# --- 2. CARGA DE DATOS ---
print("Cargando Tipo de Vía...")
gdf_vial = gpd.read_file(ruta_entrada)

# --- 3. LÓGICA DE RECLASIFICACIÓN (MAPEO) ---
# Definimos las reglas según lo que me pediste
def clasificar_vias(valor):
    valor = str(valor) # Por si hay nulos
    if valor == "Vialidad Primaria":
        return "Primaria"
    elif valor in ["Vialidad Secundaria", "Vialiadad Secundaria"]: # Incluye el error de dedo que mencionaste
        return "Secundaria"
    elif valor == "Vialidad Regional":
        return "Regional"
    elif valor == "Vialidad Federal":
        return "Federal"
    else:
        return "Otra" # Para las que no entren en tus reglas

print("Aplicando nueva categorización...")
gdf_vial['Categoría'] = gdf_vial['Clasificac'].apply(clasificar_vias)

# --- 4. LIMPIEZA DE COLUMNAS ---
print("Borrando columnas innecesarias...")
cols_a_borrar = ["Longitud_d", "Subclasifi"]
gdf_vial = gdf_vial.drop(columns=cols_a_borrar, errors='ignore')

# --- 5. GUARDADO ---
ruta_completa_salida = os.path.join(carpeta_salida, nombre_archivo_final)
gdf_vial.to_file(ruta_completa_salida, driver="GPKG")

print("-" * 30)
print(f"¡Listo! Estructura vial procesada y guardada en:\n{ruta_completa_salida}")
```
